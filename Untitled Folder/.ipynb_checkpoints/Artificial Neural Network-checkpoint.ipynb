```python
# A function to generate a random number between two given numbers, a and b
def rand(a, b):
    return (b-a)*random.random() + a
```

```python
# Make a matrix of shape a x b, filled with zero
def makeMatrix(a, b):
    return np.zeros([a, b])
```

```python
# Sigmoid function
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
# Derivative of sigmoid function
def dsigmoid(y):
    return y*(1-y)
```
